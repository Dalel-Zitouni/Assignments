```python
import pandas as pd
```

```python
df = pd.read_excel("Canada.xlsx" , header= 20 , sheet_name='Canada by Citizenship')
df.head()
```

```python
# Rename columns
df.rename(columns={
'OdName': 'Country',
'AreaName': 'Continent',
'RegName': 'Region'
}, inplace=True)


# Get year columns
Year_Columns = [col for col in df.columns if str(col).isdigit()]


# Select final columns
First_Columns = ['Country', 'Continent', 'Region', 'DevName']

df_result = df[First_Columns + Year_Columns].copy()

# Add total
df_result['Total'] = df_result[Year_Columns].sum(axis=1)
```
